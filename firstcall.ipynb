```python
import requests
import pandas as pd
import json
import os
from dotenv import load_dotenv

# safely load API key without revealing in repo
load_dotenv()
API_key = os.environ.get("API_key")

towns = ['Toms River', 'West Long Branch']

my_towns = []
maxtemp = []
mintemp = []
humidity = []
weather = []
windspeed = []

for town_names in towns:
    
    url = f'http://api.openweathermap.org/data/2.5/weather?q={town_names}&APPID={API_key}&units=imperial'
    
    r = requests.get(url)
        
    data = r.json()
    
    formatted_json = json.dumps(data, sort_keys = True, indent = 4)

    my_towns.append(data['name'])
    maxtemp.append(data['main']['temp_max'])
    mintemp.append(data['main']['temp_min'])
    humidity.append(data['main']['humidity'])
    weather.append(data['weather'][0]['description'])
    windspeed.append(data['wind']['speed'])


weather_df = pd.DataFrame()
weather_df['Names'] = my_towns
weather_df['Max_Temp'] = maxtemp
weather_df['Min_Temp'] = mintemp
weather_df['Humidity'] = humidity
weather_df['Weather'] = weather
weather_df['WindSpeed'] = windspeed

weather_df
```
